## Graficas de distribucion

### Importar librerias

```python
import seaborn as sns
import pandas as pd
```

### Importar datos

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.displot(tips['total_bill'])
```

```python
sns.displot(tips['tip'])
```

### jointplot
jointplot() basicamente representa dos funciones de distribucion

```python
sns.jointplot(x=tips['total_bill'], y=tips['tip'])
```

```python
sns.jointplot(x=tips['total_bill'], y=tips['tip'], hue = tips['sex'])
```

```python
fmri = sns.load_dataset("fmri")

# Plot the responses for different events and regions
sns.lineplot(x="timepoint", y="signal",
             hue="region", style="event",
             data=fmri)
```
