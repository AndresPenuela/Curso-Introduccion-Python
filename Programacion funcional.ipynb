# Programacion funcional
La programacion funcional se basa en tratar una funcion de la misma manera que tratamos y utilizamos una variable. Vamos a ver que quiere decir esto primero creando una funcion. La funcion va a ser muy simple y va sumar dos valores x e y. Siempre para definir una funcion debemos seguir esta plantilla.

<img src="util/plantilla_funcion.png">

```python
def suma(x,y):
    """Funcion simple que calcula la suma de dos valores x e y"""
    resultado = x + y
    return resultado
```

Desde ahora, como pasa con la variables, la funcion add se guarda en la memoria y la podemos utilizar cuando queramos.

```python
x = 2
y = 6
suma(x,y)
```

```python
suma(5,10)
```

En programacion funcional, como hemos dicho una funcion es tratada y usada igual que una variable, entonces podemos asignar una funcion, por ejemplo la funcion *suma*, a una variable, por ejemplo *a*.

```python
a = suma
```

A partir de ahora podemos utilizar *a* de igual manera que *suma*. 

```python
a(5,10)
```

Podemos decir que *a* y *suma* son variables y tambien son funciones. Y tambien podemos decir que el codigo de una funcion es un tipo de valor que toma una variable, de la misma manera que puede tomar un valor numerico (por ej: 1, 3.5, 100), una cadena de caracteres (por ej.: "Hola") o un un valor booleano (True o False).

Algo muy importante que debe tener una funcion es documentacion y comentarios que expliquen la funcion en si y las expresiones que contiene. La documentacion de la funcion viene siempre al principio de la funcion entre """. Si queremos ver esta documentacion de cualquier funcion podemos hacer click en el nombre de la funcion y pulsar Shift + Tab.
