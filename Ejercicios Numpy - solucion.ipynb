### 24. Crear una matriz de 10x10 con 1's en los bordes y 0 en el interior

```python
z = np.ones([10,10])

z[1:-1,1:-1] = 0
z
```

```python
# Una manera alternativa
z = np.zeros([10,10])

z[:,0] = 1
z[:,-1] = 1
z[0,:] = 1
z[-1,:] = 1
z
```
