```python
import pandas as pd
import numpy as np # linear algebra

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
crop = pd.read_csv('datos/Crop_recommendation.csv')
```

```python
crop.head(5)
```

```python
crop.info()
```

```python
crop.describe()
```

```python
crop_summary = pd.pivot_table(crop,index=['label'])
crop_summary.head(22)
```

```python
crop_summary = pd.pivot_table(crop,index=['label'],aggfunc='max')
crop_summary.head(22)
```

```python
crop.corr()
```

```python
sns.boxplot(x='label',y='ph',data=crop)
```

```python
fig, ax = plt.subplots(figsize=(30,15))
sns.boxplot(x='label',y='ph',data=crop)
```

```python
fig, ax = plt.subplots(figsize=(30,15))
sns.swarmplot(x='label',y='ph',data=crop)
```

```python
sns.jointplot(x="rainfall",y="humidity",data=crop,height=10,hue="label")
```

```python
sns.distplot(crop['temperature'],color="red",bins=15,hist_kws={'alpha':0.5})
```

```python
sns.pairplot(crop,hue = 'label', vars = ['K','N'])
```

```python
sns.pairplot(crop,hue = 'label', data=crop[(crop['temperature']<40) & 
                                                  (crop['rainfall']>40)])
```

```python
crop_summary_N = crop_summary.sort_values(by='N', ascending=False).reset_index()
crop_summary_N
```

```python
sns.barplot(x = crop_summary_N['N'], y = crop_summary_N['label'], data=crop_summary_N)
```
