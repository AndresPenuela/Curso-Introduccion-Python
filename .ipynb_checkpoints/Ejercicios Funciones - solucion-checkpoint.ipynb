### 3. Crea una funcion que imprima in pantalla uno a uno los elementos de esta lista 
Usa un bucle `for`

```python
messages = [ "hello world", "you are", "coding with", "Python 3" ]
```

```python
def print_list(x):
    for i in x:
        print(i)
```

```python
print_list(messages)
```

### 3. Crea una funcion que multiplique por 2 los elementos de una lista uno a uno y muestre en pantalla el resultado en cada multiplicacion.
Utiliza un bucle `for`

```python
lista = [2,5,20,50,100]
```

```python
def multiply_list(x):
    for i in x:
        print(i*2)
```

```python
multiply_list(lista)
```

### 4. Crea ahora una version de la funcion que acabas de crear pero en la que podamos poner como input el valor por le que vamos a multiplicar cada elemento de las list. Ademas, haz que la funcion multiplique por 2 por defecto si este valor no es especificado.

```python
def multiply_list(x,c=2):
    for i in x:
        print(i*c)
```

```python
multiply_list(lista,3)
```
