```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interactive
%matplotlib inline

columns=['a','b','c']
data = np.cumsum(np.random.rand(10,3),axis=1)
df = pd.DataFrame(data,columns=columns)

def g(x,y):
    plt.scatter(df[x], df[y])
    plt.show()

interactive_plot = interactive(g, x=columns, y=columns)
interactive_plot
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interactive
%matplotlib inline

x = np.linspace(0, 2 * np.pi)

slider_vals = (-1, 3, 0.1)

def g(w = 1.0):
    plt.plot(x,np.sin(w * x))
    plt.ylim((-1,1))
    plt.ylabel('eje y')
    plt.show()

interactive_plot = interactive(g, w=slider_vals)
interactive_plot
```

```python
# import ipywidgets.interact class, this class will represent a slide bar.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact
%matplotlib inline

x = np.arange(0, 2 * np.pi,0.1)
@interact(w=(-1, 3, 0.1),z=(-np.pi,np.pi,0.1))
def double_number(w=1,z=0):
    plt.plot(x,np.sin(w * x + z))
    plt.ylim((-1,1))
    plt.show()
```

```python
# import ipywidgets.interact class, this class will represent a slide bar.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ipywidgets import interact

days = np.linspace(1,101,100)
@interact(num=(1, 21, 1))
def g(num=1):
    precipitation = np.random.normal(loc = 15, scale = 2, size = (100,num))
    plt.plot(days,precipitation,linewidth=1,color ='k', alpha=0.2)
    plt.ylim((0,25))
    plt.show()
```

```python
# import ipywidgets.interact class, this class will represent a slide bar.
import numpy as np
import pandas as pd
import plotly.express as px
from ipywidgets import interact

days = np.linspace(1,101,100)
@interact(num=(1, 21, 1))
def g(num=1):
    precipitation = np.random.normal(loc = 15, scale = 2, size = (100,num))
    fig = px.line(precipitation,color_discrete_sequence=['black'],)
    fig.show()
```
